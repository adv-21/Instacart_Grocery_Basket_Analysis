```python
#defining the new function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
#applying function to the data frame
df['price_range'] = df.apply(price_label, axis=1)
```

```python
#Beginning of exercise 4.7
#using for-loop to create new busiest days column
results = []

for value in df_ords_prods_merged["orders_day_of_the_week"]:
  if value == 0 or value == 1:
    results.append("Busiest days")
  elif value == 4 or value == 3:
    results.append("Least busy")
  else:
    results.append("Regularly busy")
```

```python
#creating the new column and adding our new list to the column
df_ords_prods_merged['busiest_days'] = results
```

```python
#checking value counts to identify busiest times of day 
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
#create a for-loop to categorize the hours of the day column into the 3 categories 
result_2 = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    result_2.append("Most Orders")
  elif value in [17,8,18,19,20,7,21,22]:
    result_2.append("Average Orders")
  else:
    result_2.append("Fewest Orders")
```

```python
#adding list to new column
df_ords_prods_merged['busiest_period_of_day'] = result_2
```
