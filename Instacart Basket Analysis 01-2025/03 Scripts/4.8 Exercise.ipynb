```python
#importing libraries
import pandas as pd
import numpy as np
import os
```

```python
#importing data as dataframe
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data' , 'Prepared Data', 'orders_products_merged_newcolumns.pkl')) 
```

```python
ords_prods_merge.head(20)
```

```python
#grouping by deptartment_id to get the mean order numbers per dept.
ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

I found it interesting that overall the numbers did not change that much from the subset to the full dataframe. The subset actually gave us a very good idea of what the entire dataframe would look like 

```python
#creating max order column to show max orders for each customer
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform('max')
```

```python
#creating loyalty flag column derived from max order column value
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#grouping by loyalty flag to get the mean prices for the items they ordered
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean']})
```

It appears that in general New customers and Regular customers actually spend more/buy more expensive products than the Loyal customers.

```python
#creating average_price column to show average price of all items purchased by each customer
ords_prods_merge['average_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform('mean')
```

```python
ords_prods_merge.head(60)
```

```python
#creating spender type column derived from average_price
ords_prods_merge.loc[ords_prods_merge['average_price'] < 10, 'spender_type'] = 'Low Spender'
ords_prods_merge.loc[ords_prods_merge['average_price'] >= 10, 'spender_type'] = 'High Spender'
```

```python
#checking to see if it worked using value counts
ords_prods_merge['spender_type'].value_counts(dropna = False)
```

```python
#creating frequency column to show median days since last order for each customer
ords_prods_merge['median_days_since_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform('median')
```

```python
#checking to see if it worked
ords_prods_merge['median_days_since_order'].value_counts(dropna = False)
```

```python
#creating customer frequency column derived from median_days_since_order
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] > 20, 'customer_frequency'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_order'] <= 20) & (ords_prods_merge['median_days_since_order'] > 10), 'customer_frequency'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_order'] < 10, 'customer_frequency'] = 'Frequent customer'
```

```python
#checking to see if it worked
ords_prods_merge['customer_frequency'].value_counts()
```
